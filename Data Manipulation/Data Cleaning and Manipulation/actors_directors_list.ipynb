```python
# dependencies
import pandas as pd
```

```python
# pull main CSV to dataframe
movie_df = pd.read_csv('../Cleaned Data/cleaned_movie_data.csv')

movie_df.head()
```

```python
# create count summaries for each
billing_1 = pd.DataFrame({
    '# 1st Billing': movie_df['1st Billing'].value_counts()})
billing_2 = pd.DataFrame({
    '# 2nd Billing': movie_df['2nd Billing'].value_counts()})
billing_3 = pd.DataFrame({
    '# 3rd Billing': movie_df['3rd Billing'].value_counts()})
billing_4 = pd.DataFrame({
    '# 4th Billing': movie_df['4th Billing'].value_counts()})
billing_5 = pd.DataFrame({
    '# 5th Billing': movie_df['5th Billing'].value_counts()})
```

```python
# merge to get all columns together
actors_df = pd.merge(billing_1,
                  billing_2,
                  left_index=True,
                  right_index=True,
                  how='outer')
actors_df = pd.merge(actors_df,
                  billing_3,
                  left_index=True,
                  right_index=True,
                  how='outer')
actors_df = pd.merge(actors_df,
                  billing_4,
                  left_index=True,
                  right_index=True,
                  how='outer')
actors_df = pd.merge(actors_df,
                  billing_5,
                  left_index=True,
                  right_index=True,
                  how='outer')

# deal with nulls
actors_df = actors_df.fillna(0)

# make total column
actors_df['Total Movie Count'] = (actors_df['# 1st Billing'] +
                                  actors_df['# 2nd Billing'] +
                                  actors_df['# 3rd Billing'] +
                                  actors_df['# 4th Billing'] +
                                  actors_df['# 5th Billing'])

actors_df
```

```python
# sort and export
actors_df = actors_df.sort_values(['Total Movie Count',
                                   '# 1st Billing',
                                   '# 2nd Billing',
                                   '# 3rd Billing',
                                   '# 4th Billing',
                                   '# 5th Billing'], ascending = False)
actors_df.to_csv('../Cleaned Data/actor_counts.csv')
```

```python
# make and export directors
directors_df = pd.DataFrame({
    'Total Movie Count': movie_df['Director'].value_counts()})
directors_df.to_csv('../Cleaned Data/director_counts.csv')

directors_df
```
